```python
import requests
import datetime
import pandas as pd
```

```python
CITIES = "Paris"
COUNTRIES = "FR"
API_URL = "https://api.openweathermap.org/data/2.5/weather"
API_KEY = "fcf2e035cc53b836623ac91cccd8848d" 
```

```python
params = {"q": f"{CITIES}, {COUNTRIES}",
          "appid" : API_KEY,
          "units" : "metric"
        }
```

```python
full_url = requests.Request('GET', API_URL, params=params).prepare().url
print("requesting:", full_url)
response = requests.get(API_URL, params=params)

if response.status_code == 200:
    data = response.json()
    print(data)
else :
    print(f"ERROR {response.status_code}")
```

```python

timestamp = data['dt']
utc_time = datetime.datetime.fromtimestamp(timestamp, tz=datetime.timezone.utc)
paris_time = utc_time + datetime.timedelta(seconds=3600)
paris_time
```

```python
CITIES = ["Paris", "London", "New York", "Berlin", "Tokyo"]
COUNTRIES = ["FR", "GB", "US", "DE", "JP"]
API_URL = "https://api.openweathermap.org/data/2.5/weather"
API_KEY = "fcf2e035cc53b836623ac91cccd8848d"

# Create DataFrame
df = pd.DataFrame({"city": CITIES, "country": COUNTRIES})
df["data"] = df.apply(lambda row: requests.get(API_URL, params={"q": f"{row['city']},{row['country']}", "appid": API_KEY, "units": "metric"}).json(), axis=1)
df["temperature"] = df["data"].str["main"].str["temp"]
df["humidity"] = df["data"].str["main"].str["humidity"]
df["weather"] = df["data"].str["weather"].str[0].str["description"]
df = df.drop(columns=["data"])
```
